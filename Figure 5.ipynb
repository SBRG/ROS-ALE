## Set Up

```python
# use py file to set up workspace
from src.ros_ale_functions import *

# load data
ica = load_json_model('input/precise1k_kr.json')
ros_meta = pd.read_csv('input/ros_ale_p1k_metadata.csv', index_col = 0)
```

## Generate Figure

Final rearrangement is performed and panel L is generated in Adobe Illustrator.

```python
# layout for figure
fig = plt.figure(figsize = (6.85, 4.5), dpi = 140)
subfigs = fig.subfigures(2, 1, height_ratios = [2, 2.5])
top_axs = subfigs[0].subplots(1, 5, 
            gridspec_kw = {'width_ratios': [1, 1, 1, 1, 1],
                           'wspace': 0.4})

btm_subsubfigs = subfigs[1].subfigures(1, 2, width_ratios = [2.5, 1])
btm_axs = btm_subsubfigs[0].subplots(2, 2,
            gridspec_kw = {'wspace': 0.4, 'hspace':0.6})

# draw plots
im_bar_swarm(ica, ros_meta, 'RpoS', ax = top_axs[0])
im_bar_swarm(ica, ros_meta, 'Translation', ax = top_axs[1])
im_bar_swarm(ica, ros_meta, 'ppGpp', ax = top_axs[2])
im_bar_swarm(ica, ros_meta, 'Leucine', ax = top_axs[3])
im_bar_swarm(ica, ros_meta, 'Biotin', ax = top_axs[4])
im_bar_swarm(ica, ros_meta, 'Ribose', ax = btm_axs[0][0],
             orientation = 'horizontal', x_rot = 0)
im_bar_swarm(ica, ros_meta, 'Purine', ax = btm_axs[0][1],
             orientation = 'horizontal', x_rot = 0)
im_bar_swarm(ica, ros_meta, 'Cysteine-1', ax = btm_axs[1][0],
             orientation = 'horizontal', x_rot = 0)
im_bar_swarm(ica, ros_meta, 'Copper', ax = btm_axs[1][1],
             orientation = 'horizontal', x_rot = 0)

plt.savefig('output/fig5_raw.pdf')
```
